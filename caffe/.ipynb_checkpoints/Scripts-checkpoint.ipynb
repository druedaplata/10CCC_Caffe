###Normalization

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import cv2

img = cv2.imread('imagenes/frog.jpg',1)

b, g, r = cv2.split(img)

# create a CLAHE object (Arguments are optional).
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
r1 = clahe.apply(r)
g1 = clahe.apply(g)
b1 = clahe.apply(b)
cl1 = cv2.merge((r1,g1,b1))
img = cv2.merge((r,g,b))
plt.imshow(img)
plt.show()

plt.imshow(cl1)
plt.show()

plt.imsave("imagenes/frog2.jpg", cl1 )
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import cv2

labels = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]



for label in labels:
    lista = !ls /home/sandiego/Programas/datasets/cifar10/train-normalized/$label/
    for item in lista:
        img = cv2.imread('/home/sandiego/Programas/datasets/cifar10/train-normalized/'+label+'/'+item,1)
        b, g, r = cv2.split(img)

        # create a CLAHE object (Arguments are optional).
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(2,2))
        r1 = clahe.apply(r)
        g1 = clahe.apply(g)
        b1 = clahe.apply(b)
        cl1 = cv2.merge((r1,g1,b1))
        plt.imsave('/home/sandiego/Programas/datasets/cifar10/train-normalized/'+label+'/'+item+'-norm', cl1 )
```

###Matriz de Confusión

```python
# Cargamos CAFFE en la interfaz de python
caffe_root = '/home/sandiego/Programas/caffe/'
import sys
sys.path.insert(0, caffe_root + 'python')
import caffe
caffe.set_mode_gpu()
MODEL_FILE='/home/sandiego/Desktop/modelos/1cp1fc/deploy.prototxt'
PRETRAINED='/home/sandiego/Desktop/modelos/1cp1fc/snapshot_iter_7500.caffemodel'
net = caffe.Classifier(MODEL_FILE, PRETRAINED,raw_scale=255,image_dims=(32, 32))
    

conteo_total = []
for item in range(10):
    conteo=[]
    for j in range(10):
        conteo.append(0)
    lista1 = !grep '\b'$item'\b' /home/sandiego/Programas/datasets/train/val.txt | cut -f1 -d' '
    for file1 in lista1:      
        input_image = caffe.io.load_image("/home/sandiego/Programas/datasets/train/"+file1)
        prediction = net.predict([input_image], oversample = False) 
        conteo[prediction[0].argmax()]+=1
    conteo_total.append(conteo) 

    
    
    
import numpy
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg


conteo_clases = []
for i in range(10):
    lista2 = !grep '\b'$i'\b' /home/sandiego/Programas/datasets/train/val.txt | cut -f2 -d' '
    conteo_clases.append(len(lista2))

conteo_norm= []
for j in range(len(conteo_total)):
    aux = [float(i)/conteo_clases[j] for i in conteo_total[j]]
    conteo_norm.append(aux)
labels = ["dog","airplane","cat","truck","bird","horse","frog","ship","automobile","deer"]
plt.yticks(range(10), labels)
plt.xticks(range(10), labels, rotation='vertical')

plt.imshow(conteo_norm, interpolation='nearest')
plt.show()



```
