## 2. Elegir una Estructura de Red

<img src="imagenes/simple_net.png">

Las Redes Neuronales Convolucionales son inspiradas en la corteza visual. De esta corteza conocemos que tienen un arreglo complejo de células. Estas células son sensibles a pequeñas regiones en el campo de visión. Estas sub-regiones cubren por completo el campo visual. Las células actuan como filtros locales y están diseñadas para explotar la relación espacial presente en las imágenes naturales.


* Entrenamiento supervisado.


* Típicamente, tienen los siguientes elementos.

  * Convolutional Layer
  * Non-linearity Layer
  * Pooling Layer
  * Fully Connected Layer

## 3. Definir un solver para guiar el entrenamiento


``` python

# Con este cubrimos todas las imagenes de prueba -> batch_size * test_iter > test images
test_iter: 49
# A qué intervalos realizaremos las pruebas
test_interval: 100 

# A qué intervalo mostrar el valor de la función de costo
display: 50
# Número máximo de iteraciones
max_iter: 1000 

# Tasa de aprendizaje de la red
base_lr: 0.0001
# Política de Aprendizaje
lr_policy: "step"
stepsize: 250
gamma: 0.1
# Parámetros de regularización
momentum: 0.9
weight_decay: 0.0005

# Intervalo para guardar resultados intermedios
snapshot: 500
snapshot_prefix: "snapshot"
# Red que usaremos
net: "train_val.prototxt"
# Modo de trabajo
solver_mode: GPU # CPU o GPU
solver_type: SGD # SGD, NAG y Adagrad

```

## 4. Entrenar la red y evaluar la precisión.



````python
!/home/sandiego/Programas/caffe/build/tools/caffe train \
    --solver=/home/sandiego/10ccc_caffe/caffe/train_caffe/solver.prototxt \
    --gpu 0
````

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

# Cargamos CAFFE en la interfaz de python
caffe_root = '/home/sandiego/Programas/caffe/'
import sys
sys.path.insert(0, caffe_root + 'python')
import caffe


# Cargamos el modelo entrenado 
caffe.set_mode_gpu()
MODEL_FILE='/home/sandiego/10ccc_caffe/caffe/train_caffe/deploy.prototxt'
PRETRAINED='/home/sandiego/10ccc_caffe/caffe/snapshot_iter_1000.caffemodel'

net = caffe.Classifier(MODEL_FILE, PRETRAINED, 
                       raw_scale=255,
                       image_dims=(32, 32))

# Usamos una imagen de prueba
input_image = caffe.io.load_image('/home/sandiego/10ccc_caffe/caffe/imagenes/cat.jpg')
prediction = net.predict([input_image])
labels=['cat','truck','deer','bird','ship','automobile','frog','dog','airplane','horse']

image = mpimg.imread("/home/sandiego/10ccc_caffe/caffe/imagenes/cat.jpg")
plt.imshow(image)
plt.show()
plt.xticks(range(10), labels, rotation='vertical')
plt.bar(range(10), prediction[0], align='center')
````
